```python
df = df.rename(columns={
    'reviews.text': 'review_text',
    'reviews.rating': 'rating'
})
```

```python
df.columns
```

## Rename Columns

```python
df = df.rename(columns={
    'reviews.text': 'review_text',
    'reviews.rating': 'rating',
    'reviews.username': 'user_id'
})
```

## Clean Data

```python
df = df.drop_duplicates()
df = df.dropna(subset=['review_text', 'rating'])
```

## START FRAUD DETECTION

### Review Length:

```python
df['review_length'] = df['review_text'].apply(lambda x: len(str(x).split()))
```

### Short Review Flag:

```python
df['short_review_flag'] = df['review_length'].apply(lambda x: 1 if x < 5 else 0)
```

### Extreme Rating:

```python
df['extreme_rating'] = df['rating'].apply(lambda x: 1 if x in [1,5] else 0)
```

```python
df.columns
```

## Create Fraud Score : 

```python
df['fraud_score'] = (
    df['short_review_flag'] * 10 +
    df['extreme_rating'] * 10
)
```

## Mark Suspicious Reviews:

```python
df['is_suspicious'] = df['fraud_score'].apply(lambda x: 1 if x >= 20 else 0)
```

## Check Results

```python
df[['review_text', 'rating', 'fraud_score', 'is_suspicious']].head()
```

## Basic Analysis

### Visualization

```python
import matplotlib.pyplot as plt

df['fraud_score'].hist()
plt.title("Fraud Score Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Fraud Score Distribution:

```python
sns.histplot(df['fraud_score'], bins=20)
plt.title("Fraud Score Distribution")
plt.xlabel("Fraud Score")
plt.ylabel("Count")
plt.show()
```

### Suspicious vs Normal Reviews:

```python
sns.countplot(x='is_suspicious', data=df)
plt.title("Suspicious vs Normal Reviews")
plt.xticks([0,1], ['Normal', 'Suspicious'])
plt.show()
```

### Rating Distribution:

```python
sns.countplot(x='rating', data=df)
plt.title("Rating Distribution")
plt.show()
```

### Fraud Score vs Rating:

```python
sns.boxplot(x='rating', y='fraud_score', data=df)
plt.title("Fraud Score vs Rating")
plt.show()
```

### Short Reviews vs Fraud:

```python
sns.countplot(x='short_review_flag', hue='is_suspicious', data=df)
plt.title("Short Reviews vs Suspicious Reviews")
plt.show()
```

### Top Suspicious Users:

```python
top_users = df.groupby('user_id')['fraud_score'].mean().sort_values(ascending=False).head(10)

top_users.plot(kind='bar')
plt.title("Top Suspicious Users")
plt.ylabel("Avg Fraud Score")
plt.show()
```

### Correlation Heatmap:

```python
corr = df[['fraud_score', 'rating', 'review_length', 'short_review_flag', 'extreme_rating']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
